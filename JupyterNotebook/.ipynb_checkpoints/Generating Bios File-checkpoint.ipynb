# Automated procedure to download all the files from the website

```python
import lxml.html, io
import datetime
import re
import pandas as pd
import numpy as np
import urllib
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
# disable notification to smooth the web scraping process
options = Options()
options.add_argument("--disable-notifications")
```

```python
driver = webdriver.Chrome('/Applications/chromedriver_mac_arm64/chromedriver', chrome_options=options)
driver.implicitly_wait(10)
driver.get("https://bioguide.congress.gov/search")

try:
    myElem = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'search-downloads')))
    print("Page is ready!")
    
except TimeoutException:
    print("Loading took too much time!")
    driver.quit()

time.sleep(3)    
download_button = chrome.find_element(By.XPATH, "//*[@id=\"search-downloads\"]/div/div")
download_button.click()
bulkdata_button = chrome.find_element(By.ID, "Download bulk data in json format")
bulkdata_button.click()
allprofiles_button = chrome.find_element(By.XPATH, "//*[@id=\"search-downloads\"]/div[1]/div/div/div/div[2]/div/div")
allprofiles_button.click()
time.sleep(30)


driver.quit()
```

# Contenating bios

```python
import json
import os
import csv
import pandas as pd
from pathlib import Path
from pandas.io.json import json_normalize
```

```python
os.getcwd()
cwd = Path("/Users/elainewei/Dropbox/Berkeley/Research/Malmendier/PoliticianVoting")
```

```python
def Json2df(filename):
    """
    Takes in a json file and convert it into a row of dataframe.
    """
    with open(f"{cwd}/raw/BioguideProfiles/{filename}") as json_file:
        biodict = json.load(json_file) # change into dictionary
        df = pd.json_normalize(biodict) # change into dataframe
    return df
```

```python
# example

df1 = Json2df("L000226.json")
df2 = Json2df("M000777.json")
df = pd.concat([df1,df2]).reset_index(drop = True)
display(df)
```

```python
def GenDF():
    """
    Contenates the data of each politician.
    """
    filenames = os.listdir(f"{cwd}/raw/BioguideProfiles")
    
    df = pd.DataFrame()
    
    for filename in filenames:
        new_df = Json2df(filename)
        df = pd.concat([df,new_df]).reset_index(drop=True)
        print(f"{filename} done))
    return df
```

```python
# Generating the csv file containing all the information of the politicians
bioguide_df = GenDF()
bioguide_df.to_csv(f"{cwd}/bioguide.csv")
```
